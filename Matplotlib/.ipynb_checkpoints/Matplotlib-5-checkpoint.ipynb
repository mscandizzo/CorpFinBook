### Two-Dimensional Plotting

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import warnings; warnings.simplefilter('ignore')
import seaborn as sns; sns.set()
%matplotlib inline
```

### One-Dimensional Data Set

```python
spy = pd.read_csv('SPY.csv',index_col=0)
spy.index = pd.to_datetime(spy.index)
```

```python
plt.figure(figsize=(12,8))
plt.plot(spy['Adj Close'],'--',lw=2,marker='o',color='blue',markerfacecolor='green', markeredgecolor='k')
plt.xticks(fontsize=10, rotation=45)
plt.title('SPY 2019 Historical Evolution',fontsize=14)
plt.ylabel('SPY Price',fontsize=12)
plt.xlabel('Dates',fontsize=12)
plt.grid(True)  # adds a grid
plt.axis('tight')  ;# adjusts the axis ranges
plt.ylim(220,340);
plt.show();
```

### Two-Dimensional Data Set

```python
plt.figure(figsize=(12,8))
plt.plot(spy['Close'],'--',lw=2,marker='o',color='blue',markerfacecolor='royalblue', 
         markeredgecolor='k',label='Close')
plt.plot(spy['High']+10,'--',lw=2,marker='2',color='green',markerfacecolor='lime',label='High')
plt.plot(spy['Low']-10,'--',lw=2,marker='1',color='red',markerfacecolor='darkred',label='Close')
plt.xticks(fontsize=10, rotation=45)
plt.title('SPY 2019 Historical Evolution',fontsize=14)
plt.ylabel('SPY Price',fontsize=12)
plt.xlabel('Dates',fontsize=12)
plt.grid(True)  # adds a grid
plt.axis('tight') # adjusts the axis ranges
plt.ylim(220,350)
plt.legend(loc=2)
plt.show();
```

```python
fig, ax1 = plt.subplots(figsize=(12,8))
plt.plot(spy['Close'],'--',lw=2,marker='o',color='blue',markerfacecolor='royalblue', 
         markeredgecolor='k',label='Close')
plt.grid(True)
plt.legend(loc=9,fontsize=12)
plt.axis('tight')
plt.xlabel('Dates',fontsize=12)
plt.ylabel('SPY Price',fontsize=12)
plt.xticks(fontsize=10, rotation=45)
plt.title('SPY 2019 Historical Closing price and Volume',fontsize=16)
ax2 = ax1.twinx()
plt.plot(spy['Volume'],'--',lw=2,marker=',',color='orange',markerfacecolor='darkorange', 
         markeredgecolor='k',label='Volume')
plt.legend(loc=8,fontsize=12)
plt.ylabel('SPY Volume',fontsize=12)
plt.show();
```

### Other Plot Styles

```python
plt.figure(figsize=(12, 8))
plt.scatter(spy['Close'].values,spy['Volume'].values, marker='8',color='royalblue', 
         edgecolors='black')
plt.grid(True)
plt.xlabel('SPY Close Price',fontsize=12)
plt.ylabel('SPY Volume',fontsize=12)
plt.title('SPY Close-Volume Scatter Plot',fontsize=16);
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(spy['Close'].values,spy['Volume'].values, c=spy['Close'].values, marker='o')
plt.colorbar()
plt.grid(True)
plt.xlabel('SPY Close Price',fontsize=12)
plt.ylabel('SPY Volume',fontsize=12)
plt.title('SPY Close-Volume Scatter Plot',fontsize=16);
```

```python
plt.figure(figsize=(12, 8))
plt.hist(spy[['High','Low']].values, label=['High', 'Close'], bins=25, edgecolor='darkblue',color=['royalblue','red'])
plt.grid(True)
plt.legend(loc=0,fontsize=14)
plt.xlabel('Values',fontsize=12)
plt.ylabel('Frequency',fontsize=12)
plt.title('SPY High & Low Prices Histogram',fontsize=16);
```

```python
plt.figure(figsize=(12, 8))
plt.hist(spy[['High','Low']].values, label=['High', 'Close'], bins=25, edgecolor='darkblue',color=['royalblue','red'],stacked=True)
plt.grid(True)
plt.legend(loc=0,fontsize=14)
plt.xlabel('Values',fontsize=12)
plt.ylabel('Frequency',fontsize=12)
plt.title('SPY High & Low Prices Histogram - Stacked Version',fontsize=16);
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
green_diamond = dict(markerfacecolor='g', marker='D')

ax = plt.boxplot(spy[['High','Low']].values,flierprops=green_diamond, labels=['High','Low'],
            notch=True, patch_artist=True)
plt.grid(True)
colors = ['royalblue', 'red']
for patch, color in zip(ax['boxes'], colors):
    patch.set_facecolor(color)
    
plt.xlabel('Data sets',fontsize=12)
plt.ylabel('Values',fontsize=12)
plt.title('SPY High & Low Boxplot',fontsize=16)
plt.show();
```

```python
from matplotlib.patches import Polygon
def func(x):
    return 0.5 * np.exp(x) + 1

a, b = 0.5, 1.5  # integral limits
x = np.linspace(0, 2)
y = func(x)

fig, ax = plt.subplots(figsize=(7, 5))
plt.plot(x, y, 'b', linewidth=2)
plt.ylim(ymin=0)

# Illustrate the integral value, i.e. the area under the function
# between lower and upper limit
Ix = np.linspace(a, b)
Iy = func(Ix)
verts = [(a, 0)] + list(zip(Ix, Iy)) + [(b, 0)]
poly = Polygon(verts, facecolor='0.7', edgecolor='0.5')
ax.add_patch(poly)

plt.text(0.5 * (a + b), 1, r"$\int_a^b f(x)\mathrm{d}x$",
         horizontalalignment='center', fontsize=20)

plt.figtext(0.9, 0.075, '$x$')
plt.figtext(0.075, 0.9, '$f(x)$')

ax.set_xticks((a, b))
ax.set_xticklabels(('$a$', '$b$'))
ax.set_yticks([func(a), func(b)])
ax.set_yticklabels(('$f(a)$', '$f(b)$'))
plt.grid(True)
```

## 3d Plotting

```python
StrikePrice = np.linspace(250, 450, 100)
TimetoMaturity = np.linspace(.5, 2.5, 100)
StrikePrice, TimetoMaturity = np.meshgrid(StrikePrice, TimetoMaturity)
```

```python
ImpliedVol = (StrikePrice - 100) ** 2 / (100 * StrikePrice) / TimetoMaturity
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig1 = plt.figure(figsize=(16, 8))
ax = fig1.gca(projection='3d')

surf = ax.plot_surface(StrikePrice, TimetoMaturity, ImpliedVol,
                       rstride=2, cstride=2,cmap=plt.cm.coolwarm, 
                       linewidth=0.5,antialiased=True)

ax.set_title('Volatility Smile',fontsize=16)
ax.set_xlabel('Strike',fontsize=14)
ax.set_ylabel('Time to Maturity',fontsize=14)
ax.set_zlabel('implied volatility',fontsize=14)

fig1.colorbar(surf, shrink=0.8, aspect=8);
```
