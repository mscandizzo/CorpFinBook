## Monte Carlo Simulation
-----------------

```python
import numpy as np
np.random.seed(1000)
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def simul(S0, mu , std, T, Y, Simul): # Generates Monte Carlo paths for Geometric Brownian motion
    dt = float(T) / Y
    paths = np.zeros((Y+1 , Simul), np.float64)
    paths[0] = S0
    for t in range(1, Y + 1):
        rand = np.random.standard_normal(Simul)
        rand = (rand - rand.mean())/ rand.std()
        paths[t] = paths[t - 1] * np.exp((mu - 0.5 * std **2) *dt + std * np.sqrt(dt) * rand)
    return paths
   
```

```python
S0 = 36.24 # Initial Revenue data point
mu = 0.07 # drift 
std = 0.3 # standard deviation
T = 1.0 # time step
Y = 10 # number of periods
Simul = 30 # number of simulations
```

```python
paths = simul(S0, mu, std, T, Y, Simul)
```

```python
MC = pd.DataFrame(paths)
```

```python
MC
```

```python
MC['Average'] = MC.mean(axis=1)
```
