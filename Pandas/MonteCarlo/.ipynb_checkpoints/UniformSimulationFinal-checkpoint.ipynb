```python
import pandas as pd
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt
import seaborn as sns; sns.set
from matplotlib.ticker import FuncFormatter
from collections import namedtuple
%matplotlib inline
```

from ipyparallel import Client

```python
class UniSetting(object):
    '''Class model to run Monte Carlo simulation using Uniform distribution
    Parameters
    ==========
    gp: float
       Gross profit 
    writeoff: float
       product waste which is a negative result at branch level
    daysales: float
       Average daily amount in sales based on initial area survey
    upvar: float
       Upward random positive deviation from average value
    downvar: float
       Upward random positive deviation from average value
    months: integer
       Number of months to run simulation
    monthcost: float
       Monthly total cost including Rent, Salaries, Utitilities, etc.
    costgrowth: float
       Monthly expected growth in cost for the life of the simulation
    startcost: float
       Initial cost in fit out and equipment
    salesincrease: float
       Monthly expected growth in sales
    simul: integer
       Number of simulations to run
    '''
    def __init__(self, gp = 0.34, writeoff = 0.04, daysales = 15000, upvar = 0.3 , downvar = 0.3,
                 months = 120, monthcost = 71181 , costgrowth = 0.005, startcost = 1200000, 
                 salesincrease = 0.005, simul = 100, seed = 20, annual = 12, days = 30):
        self.gp = gp
        self.writeoff = writeoff
        self.daysales = daysales
        self.upvar = upvar
        self.downvar = downvar
        self.months = months
        self.monthcost = monthcost
        self.costgrowth = costgrowth
        self.startcost = startcost
        self.salesincrease = salesincrease
        self.simul = simul
        self.seed = seed
        self.annual = annual
        self.days = days
        self.Pricing = self.Prices_Simul()
        self.RevM = self.Revenue()
        self.CosM = self.Costing()
        self.Profit = self.Profit()
        self.Results = self.Results()
        
    '''
    Methods
    =======
    Uniform_Simul:
        The simulation uses the average daily sales which grow at the salesincrease rate.
        the initial start value value at t = 0 is equal to the sum of fitout cost + equipment cost
        for each point in time a range is crate which is define as:
        lower boundary: daysales - downvar
        upper boundary: daysales + upvar
        a random value within that range it is use to simulate
        the random monthly amount - the monthly cost is use to calculate the net monthly result   
    '''
        
    def Prices_Simul(self, rows=None, columns=None, initial=None, up=None,
                     down=None, increase=None, seed=None):
        rows = rows or self.months
        columns = columns or self.simul
        initial = initial or self.daysales
        up = up or self.upvar
        down = down or self.downvar
        increase = increase or self.salesincrease
        seed = seed or self.seed

        Price = np.zeros((rows + 1, columns))
        Price[0] = initial
        # Set seed number in order to lock simulation values
        npr.seed(seed)
        for t in range(1, rows + 1):
            rmin = initial * (1 - down)
            rmax = initial * (1 + up)
            Price[t] = npr.uniform(rmin, rmax, columns)
            initial *= (1 + increase)
        self.Price = pd.DataFrame(Price)
        return self.Price

    def Costing(self, rows=None, mcost=None, gcost=None):
        rows = rows or self.months
        mcost = mcost or self.monthcost
        gcost = gcost or self.costgrowth

        cost = np.ones(rows + 1) * mcost
        costg = [((1 + gcost) ** i) * x for i, x in enumerate(cost)]
        costg[0] = 0
        costg = np.array(costg).reshape(rows + 1, 1)
        self.costg = pd.DataFrame(costg)
        return self.costg

    def Revenue(self, days = None ,rows=None, columns=None, gp=None, writeoff=None, startcost=None):
        rows = rows or self.months
        columns = columns or self.simul
        gp = gp or self.gp
        writeoff = writeoff or self.writeoff
        startcost = startcost or self.startcost
        days = days or self.days
        
        PriceM = np.zeros((rows + 1, columns))
        PriceM = self.Pricing
        self.PriceM = pd.DataFrame(PriceM)
        NetMonth = (gp - writeoff) * self.days
        self.PriceM = pd.DataFrame(self.PriceM.values * NetMonth)
        self.PriceM.iloc[0] = - self.startcost
        return self.PriceM

    def Profit(self, revenue=None, cost=None):
        self.PandL = self.RevM.values - self.CosM.values
        self.PandL = pd.DataFrame(self.PandL)
        return self.PandL

    def Results(self):
        Outcome = namedtuple('Results',
                             'EndAmount AvPLPath PLPaths AvPrice PricePaths Return')
        self.EndAmount = self.Profit.cumsum(axis = 0).iloc[-1].mean()#It takes the final average value of the cumsum of all path
        self.AvPLPath = pd.DataFrame(self.Profit.cumsum(axis = 0).mean(axis = 1))# Average P&L path 
        self.PLPaths = self.Profit.cumsum(axis = 0)# All simulated P&L paths
        
        # Price simulations paths
        self.AvPrice = pd.DataFrame(self.Pricing.mean(axis = 1))# Average Price path 
        self.PricePaths = self.Pricing# All simulated Price paths
        
        # Calculation of final return assuming final cummulative result over initial investment
        self.returns = ((self.EndAmount / self.startcost) **(1/(self.months/12)) - 1) * 100
        #self.returns = returns
        
        return Outcome(EndAmount = self.EndAmount,
                       AvPLPath = self.AvPLPath,
                       PLPaths = self.PLPaths,
                       AvPrice = self.AvPrice,
                       PricePaths = self.PricePaths,
                       Return = self.returns)
    def PLplotting(self):
        fig = plt.figure(figsize=(14,6))
        
        axes1 = fig.add_subplot(121)
        axes1.plot(self.AvPLPath)
        axes1.get_yaxis().set_major_formatter(FuncFormatter(lambda x, p: format(int(x), ',')))
        axes1.set_xlabel('Months',fontsize = 12)
        axes1.set_ylabel('P&L',fontsize = 12)
        axes1.set_title('Avg.P&L, Annualized Compounded Return: {:1.2f} %'.format(self.returns),fontsize = 16)
        axes1.axhline(0 ,linewidth = 3, color = 'r', linestyle =":", marker = "o")
        
        axes2 = fig.add_subplot(122, sharex = axes1, sharey = axes1)
        axes2.plot(self.PLPaths)
        axes2.set_xlabel('Months',fontsize = 12)
        axes2.set_ylabel('P&L',fontsize = 12)
        axes2.set_title('P&L Simulations',fontsize = 16)
        axes2.axhline(0 ,linewidth = 3, color = 'r', linestyle =":", marker = "o")
        
        plt.suptitle('Monte Carlo Price Simulations', fontsize = 20)
        plt.show()
    
    def Priceplotting(self):
        fig = plt.figure(figsize=(14,6))
        
        axes1 = fig.add_subplot(121)
        axes1.plot(self.AvPrice)
        axes1.get_yaxis().set_major_formatter(FuncFormatter(lambda x, p: format(int(x), ',')))
        axes1.set_xlabel('Months',fontsize = 12)
        axes1.set_ylabel('Daily Sales',fontsize = 12)
        axes1.set_title('Daily Sales Average Path',fontsize = 16)
        
        axes2 = fig.add_subplot(122, sharex = axes1, sharey = axes1)
        axes2.plot(self.PricePaths)
        axes2.set_xlabel('Months',fontsize = 12)
        axes2.set_ylabel('Daily Sales',fontsize = 12)
        axes2.set_title('Daily Sales Simulations',fontsize = 16)
        
        plt.suptitle('Monte Carlo Price Simulations', fontsize = 20)
        plt.show()
    
    def DownloadResults(self):
        writer = pd.ExcelWriter(r'C:\Users\maria\OneDrive - Scandizzo & Partner, LLC\SPartners\Python\simul1.xlsx',engine='xlsxwriter')
        self.AvPLPath.to_excel(writer, sheet_name ='AvPLPath')
        self.PLPaths.to_excel(writer, sheet_name ='PLPaths')
        self.AvPrice.to_excel(writer, sheet_name ='AvPrice')
        self.PricePaths.to_excel(writer, sheet_name ='PricePaths')
        self.costg.to_excel(writer, sheet_name ='costg')
        self.PriceM.to_excel(writer, sheet_name ='Revenue')
        writer.save()
        print('The end Amount is AED {:1.2f} after {:1.0f} years'.format(self.EndAmount, self.months/self.annual))
        
        
```
